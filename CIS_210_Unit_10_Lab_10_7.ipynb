```python
# zyBooks solution
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor

diamonds = pd.read_csv('diamonds.csv')

diamonds = diamonds.sample(n=800, random_state=10)

# Create a dataframe X containing all the features except cut, color, clarity, and price
X = diamonds.drop(columns=['cut', 'color', 'clarity', 'price'])
# Create a dataframe y containing the feature price
y = diamonds[['price']]

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=123)

# Initialize a multilayer perceptron regressor with two hidden layers with 50 nodes each,
# the 'identity' activation function, max_iter=500, and random_state=123
mlpModel = MLPRegressor(hidden_layer_sizes=[50,50], activation='identity', max_iter=500, random_state=123)
# Fit the model
mlpModel.fit(Xtrain, np.ravel(ytrain))

# Print the price predictions and actual prices for the first five rows of the dataframe
print("Price predictions:", mlpModel.predict(Xtrain[0:5]))
print("Actual prices: \n", ytrain[0:5])

# Print the score for the training data
print("Score for the training data: ", round(mlpModel.score(Xtrain, ytrain), 4))
# Print the score for the testing data
print("Score for the testing data: ", round(mlpModel.score(Xtest, ytest), 4))
```

```python
# Claude solution
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split

# Load the dataset
diamonds = pd.read_csv('diamonds.csv')

# Create feature dataframe X (all features except cut, color, clarity, and price)
X = diamonds.drop(['cut', 'color', 'clarity', 'price'], axis=1)

# Create target dataframe y containing price
y = diamonds['price']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=123
)

# Initialize and fit multilayer perceptron regressor
mlp_model = MLPRegressor(
    hidden_layer_sizes=(50, 50),
    activation='identity',
    max_iter=500,
    random_state=123
)

mlp_model.fit(X_train, y_train)

# Print price predictions for the first five rows of training data
predictions_first_five = mlp_model.predict(X_train[:5])
print(f"Price predictions: {predictions_first_five}")

# Print actual prices for the first five rows of training data
print("Actual prices: ")
print(y_train[:5].to_frame())

# Calculate and print R-squared scores
train_score = mlp_model.score(X_train, y_train)
test_score = mlp_model.score(X_test, y_test)

print(f"Score for the training data:  {round(train_score, 4)}")
print(f"Score for the testing data:  {round(test_score, 4)}")
```
