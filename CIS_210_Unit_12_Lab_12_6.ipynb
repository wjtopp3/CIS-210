```python
# zyBook Solution
# Load the necessary libraries
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.manifold import MDS, Isomap, TSNE
from sklearn.preprocessing import MinMaxScaler

# Silence warning
import warnings
warnings.filterwarnings('ignore')

concrete = pd.read_csv('concrete.csv').sample(n=100, random_state=123)

# Input the number of principal components
nComp = int(input())

X = concrete.drop(['strength'], axis=1)
y = concrete[['strength']]

scaler = MinMaxScaler()

# Fit an MDS model with a user defined number of components and random_state=42
# to scaled input data
mds = MDS(n_components=nComp, random_state=42)
pipeline_mds = Pipeline(steps=[('scaler',scaler),('mds',mds)])
X_transform_mds = pipeline_mds.fit_transform(X)

# Find the data points in lower-dimensional space
mdsDataPoints = mds.embedding_

# Display the data points in lower-dimensional space
print('MDS:\n', pd.DataFrame(mdsDataPoints))

# Fit an isomap model with a user defined number of components and n_neighbors=12
# to scaled input data
isomap = Isomap(n_components=nComp, n_neighbors=12)
pipeline_isomap = Pipeline(steps=[('scaler',scaler),('isomap',isomap)])
X_transform_isomap = pipeline_isomap.fit_transform(X)

# Find the data points in lower-dimensional space
isomapDataPoints = isomap.embedding_

# Display the data points in lower-dimensional space
print('Isomap:\n', pd.DataFrame(isomapDataPoints))

# Fit a t-SNE model with a user defined number of components, perplexity=20.0, and random_state=42
# to scaled input data
tsne = TSNE(n_components=nComp, perplexity=20.0, random_state=42)
pipeline_tsne = Pipeline(steps=[('scaler',scaler),('tsne',tsne)])
X_transform_tsne = pipeline_tsne.fit_transform(X)

# Find the data points in lower-dimensional space
tsneDataPoints = tsne.embedding_

# Display the data points in lower-dimensional space
print('t-SNE:\n', pd.DataFrame(tsneDataPoints))
```

```python
# Claude Solution
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import MDS, Isomap, TSNE

# Read the dataset
df = pd.read_csv('concrete.csv')

# Get the number of components from user input
n_components = int(input())

# Select the input features (8 features, excluding the target variable)
# Assuming the last column is the target (compressive strength)
X = df.iloc[:, :-1].values

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- MDS (Multidimensional Scaling) ---
mds = MDS(n_components=n_components, random_state=42)
X_mds = mds.fit_transform(X_scaled)
df_mds = pd.DataFrame(X_mds)

print("MDS:")
print(df_mds)

# --- Isomap (Isometric Mapping) ---
isomap = Isomap(n_components=n_components, n_neighbors=12)
X_isomap = isomap.fit_transform(X_scaled)
df_isomap = pd.DataFrame(X_isomap)

print("Isomap:")
print(df_isomap)

# --- t-SNE (t-distributed Stochastic Neighbor Embedding) ---
# Note: t-SNE may produce error with 4+ components
tsne = TSNE(n_components=n_components, perplexity=20.0, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)
df_tsne = pd.DataFrame(X_tsne)

print("t-SNE:")
print(df_tsne)
```
