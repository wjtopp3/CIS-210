```python
# zyBooks solution
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

mpg = pd.read_csv('mpg.csv')
plt.rcParams['figure.dpi'] = 150

# Create a dataframe X containing cylinders, weight, and mpg
X = mpg[['cylinders', 'weight', 'mpg']]

# Create a dataframe y containing origin
y = mpg[['origin']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)

# Initialize and fit the tree
DTC = DecisionTreeClassifier(max_leaf_nodes=6)
DTC.fit(X_train,y_train)

# Print the text summary of the tree
DTC_tree = export_text(DTC)
print(DTC_tree)

# Make predictions for the test data
y_pred=DTC.predict(X_test)

# Create and plot a confusion matrix
cm = confusion_matrix(y_test,y_pred)
ConfusionMatrixDisplay(cm).plot()
plt.savefig('confMatrix.png')
```

```python
# Claude solution
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn import tree

# Read the dataset
df = pd.read_csv('mpg.csv')

# Create dataframe X with features cylinders, weight, and mpg
X = df[['cylinders', 'weight', 'mpg']]

# Create dataframe y with the target feature origin
y = df['origin']

# Split the data into training and testing sets (default 75-25 split)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=123
)

# Initialize and fit the classification tree
clf = DecisionTreeClassifier(max_leaf_nodes=6, random_state=123)
clf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = clf.predict(X_test)

# Print the tree structure
tree_rules = tree.export_text(clf, feature_names=list(X.columns))
print(tree_rules)

# Create and print the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
