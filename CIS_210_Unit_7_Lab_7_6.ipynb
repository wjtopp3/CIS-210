```python
# zyBooks solution
# Import the necessary packages
import numpy as np
import pandas as pd
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

mpg = pd.read_csv('mpg.csv')

# User-defined input feature
X = mpg[[input()]]

# Output feature: mpg
y = mpg[['mpg']]

# Create training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)

# Scale the input features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
y_train = scaler.fit_transform(y_train)
y_test = scaler.transform(y_test)

# Initialize and fit a linear svr model to training data with epsilon = 0.2
eps = 0.2
svr_lin = SVR(kernel='linear', epsilon=eps)
svr_lin.fit(np.reshape(X_train,(-1,1)), np.ravel(y_train))

# Initialize and fit an svr model using a poly kernel with epsilon=0.2, C=0.5, and gamma=0.7
svr_poly = SVR(kernel='poly', epsilon=eps, C=0.5, gamma=0.7)
svr_poly.fit(np.reshape(X_train,(-1,1)), np.ravel(y_train))

# Initialize and fit an svr model using an rbf kernel with epsilon=0.2, C=0.5, and gamma=0.7
svr_rbf = SVR(kernel='rbf', epsilon=eps, C=0.5, gamma=0.7)
svr_rbf.fit(np.reshape(X_train,(-1,1)), np.ravel(y_train))

# Print the coefficients of determination for each model
lin_score = svr_lin.score(X_test, y_test)
print('Linear model:', np.round(lin_score, 3))

poly_score = svr_poly.score(X_test, y_test)
print('Polynomial model:', np.round(poly_score, 3))

rbf_score = svr_rbf.score(X_test, y_test)
print('RBF model:', np.round(rbf_score, 3))
```

```python
# Claude solution
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR

# Read the dataset
df = pd.read_csv('mpg.csv')

# Get user input for the feature to use
input_feature = input()

# Separate features and target
X = df[[input_feature]]  # Feature selected by user
y = df['mpg']  # Target variable

# Split the data: 75% training, 25% testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=123
)

# Initialize and fit Linear SVR
linear_svr = SVR(kernel='linear', epsilon=0.2)
linear_svr.fit(X_train, y_train)

# Initialize and fit Polynomial SVR
poly_svr = SVR(kernel='poly', epsilon=0.2, C=0.5, gamma=0.7)
poly_svr.fit(X_train, y_train)

# Initialize and fit RBF SVR
rbf_svr = SVR(kernel='rbf', epsilon=0.2, C=0.5, gamma=0.7)
rbf_svr.fit(X_train, y_train)

# Calculate and print RÂ² scores (coefficient of determination)
linear_score = linear_svr.score(X_test, y_test)
poly_score = poly_svr.score(X_test, y_test)
rbf_score = rbf_svr.score(X_test, y_test)

print(f"Linear model: {linear_score:.3f}")
print(f"Polynomial model: {poly_score:.3f}")
print(f"RBF model: {rbf_score:.3f}")
```
