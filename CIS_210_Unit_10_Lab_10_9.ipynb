```python
# zyBooks solution
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import os

# Set the backend using tensorflow
os.environ['KERAS_BACKEND'] = 'tensorflow'
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '1'

# The backend must be set before importing keras, not after
import keras
keras.utils.set_random_seed(812)

df = pd.read_csv('diamonds.csv')
diamond_sample = df.sample(1000, random_state=12)

X = diamond_sample.drop(columns=['cut', 'color', 'clarity', 'price'])
y = diamond_sample[['price']]

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3)

# Define the model structure using keras.Sequential

model = keras.Sequential(
    [
        # Input layer with shape=(6, )
        keras.layers.Input(shape=(6, )),
        # Hidden layer 1: 256 nodes, relu activation
        keras.layers.Dense(256, activation='relu'),
        # Hidden layer 2: 128 nodes, linear activation
        keras.layers.Dense(128, activation='linear'),
        # Hidden layer 3: 64 nodes, linear activation
        keras.layers.Dense(64, activation='linear'),
        # Output layer: 1 node for regression, linear activation
        keras.layers.Dense(1, activation='linear'),
    ]
)

print(model.summary(line_length=80))


# Specify training choices using the compile method, with optimizer='Adam', loss='MeanSquaredError',
# metrics='mse'
model.compile(
    optimizer='Adam',  # Optimizer
    # Loss function to minimize
    loss='MeanSquaredError',
    # List of metrics to monitor
    metrics=['mse'],
)

# Train the model with a batch size of 100, 5 epochs, validation_split=0.1, and verbose=0
training = model.fit(Xtrain, ytrain, batch_size=100, epochs=5, validation_split=0.1, verbose=0)

predictions = model.predict(Xtest[:3], verbose=0)
print('Predictions:', predictions.round(3))
print('Actual values:', ytest[:3])
```

```python
# Claude solution
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split

# Set the backend to tensorflow
keras.backend.set_floatx('float32')

# Load the dataset
diamonds = pd.read_csv('diamonds.csv')

# Create feature dataframe X (all features except cut, color, clarity, and price)
X = diamonds.drop(['cut', 'color', 'clarity', 'price'], axis=1)

# Create target dataframe y containing price
y = diamonds['price']

# Split the data into training and testing sets
Xtrain, Xtest, ytrain, ytest = train_test_split(
    X, y, random_state=42
)

# Define the model structure using keras.Sequential
model = keras.Sequential([
    keras.layers.Dense(256, activation='relu', input_shape=(6,)),
    keras.layers.Dense(128, activation='linear'),
    keras.layers.Dense(64, activation='linear'),
    keras.layers.Dense(1, activation='linear')
])

# Compile the model
model.compile(
    optimizer='Adam',
    loss='MeanSquaredError',
    metrics=['mse']
)

# Train the model
model.fit(
    Xtrain,
    ytrain,
    batch_size=100,
    epochs=5,
    validation_split=0.1,
    verbose=0
)

# Print model summary
print(model.summary())

# Make predictions on the first 3 rows of test data
predictions = model.predict(Xtest[:3], verbose=0)
print(f"Predictions: {predictions}")

# Print actual values
print(f"Actual values: {ytest[:3].to_frame()}")
```
