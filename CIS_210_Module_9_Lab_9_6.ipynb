```python
# zyBooks Solution
import numpy as np
import pandas as pd
from sklearn.ensemble import BaggingRegressor

df = pd.read_csv('msleep_clean.csv')

# Create a dataframe X containing the features awake, brainwt, and bodywt, in that order
X = df[['awake', 'brainwt', 'bodywt']]

# Create a dataframe y containing sleep_rem
y = df[['sleep_rem']]

# Initialize and fit bagging regressor with 30 base estimators, a random state of 10, and oob_score=True
sleepModel = BaggingRegressor(n_estimators=30, random_state=10, oob_score=True)
sleepModel.fit(X, np.ravel(y))

# Calculate out-of-bag accuracy
print(np.round(sleepModel.oob_score_, 4))

# Calculate predictions from out-of-bag estimate
print(np.round(sleepModel.oob_prediction_, 4))
```

```python
# Claude solution
import pandas as pd
from sklearn.ensemble import BaggingRegressor

# Load the dataset
msleep_clean = pd.read_csv('msleep_clean.csv')

# Create feature dataframe X with awake, brainwt, and bodywt
X = msleep_clean[['awake', 'brainwt', 'bodywt']]

# Create target dataframe y with sleep_rem
y = msleep_clean['sleep_rem']

# Initialize and fit bagging regressor
bagging_model = BaggingRegressor(
    n_estimators=30,
    random_state=10,
    oob_score=True
)

bagging_model.fit(X, y)

# Print OOB score (rounded to 4 decimal places)
print(round(bagging_model.oob_score_, 4))

# Print predictions
predictions = bagging_model.predict(X)
print(predictions)
```
