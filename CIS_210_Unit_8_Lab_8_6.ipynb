```python
# zyBooks solution
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor, export_text
from sklearn.model_selection import train_test_split

sleep = pd.read_csv('msleep_clean.csv')

# Create a dataframe X containing the features awake, brainwt, and bodywt
X = sleep[['awake', 'brainwt', 'bodywt']]


# Output feature: sleep_rem
y = sleep[['sleep_rem']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Initialize the model with max_depth=3, ccp_alpha=0.02, and random_state=123
DTRModel = DecisionTreeRegressor(max_depth=3, ccp_alpha=0.02, random_state=123)

# Fit the model
DTRModel.fit(X_train, y_train)

# Print the R-squared value for the testing set
print(DTRModel.score(X_test, y_test))

# Print text summary of tree
DTR_tree = export_text(DTRModel)
print(DTR_tree)
```

```python
# Claude solution
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree

# Read the dataset
df = pd.read_csv('msleep_clean.csv')

# Create dataframe X with features awake, brainwt, and bodywt (in that order)
X = df[['awake', 'brainwt', 'bodywt']]

# Create dataframe y with the target feature sleep_rem
y = df['sleep_rem']

# Split the data into training and testing sets (default 75-25 split)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=123
)

# Initialize and fit the regression tree
regressor = DecisionTreeRegressor(
    max_depth=3,
    ccp_alpha=0.02,
    random_state=123
)
regressor.fit(X_train, y_train)

# Calculate and print the R-squared value for the testing set
r2_score = regressor.score(X_test, y_test)
print(r2_score)

# Print the tree structure
tree_rules = tree.export_text(regressor, feature_names=list(X.columns))
print(tree_rules)
```
