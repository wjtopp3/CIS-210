```python
# Claude's solution
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer, KNNImputer, IterativeImputer
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
msleep = pd.read_csv('msleep.csv')

# Create dataframes with selected features
features = ['sleep_rem', 'sleep_cycle', 'awake', 'brainwt']
mammalSimple = msleep[features].copy()
mammalKnn = msleep[features].copy()
mammalIter = msleep[features].copy()

# Display dataframe with missing data (showing first 5 rows with some missing values)
print("Dataframe with missing data: ")
print(mammalSimple.head())

# 1. SimpleImputer - Fill with most frequent value
simple_imputer = SimpleImputer(strategy='most_frequent')
simple_imputed = simple_imputer.fit_transform(mammalSimple)
mammalSimple_filled = pd.DataFrame(simple_imputed, columns=features)

print("Dataframe filled with SimpleImputer: ")
print(mammalSimple_filled.head())

# 2. KNNImputer - Use 4 neighbors with distance weights
# First scale the data for KNN imputation
scaler = MinMaxScaler()
mammalScaled = pd.DataFrame(
    scaler.fit_transform(mammalKnn),
    columns=['scaled_' + col for col in features]
)

knn_imputer = KNNImputer(n_neighbors=4, weights='distance')
knn_imputed = knn_imputer.fit_transform(mammalScaled)
mammalKnn_filled = pd.DataFrame(
    knn_imputed,
    columns=['scaled_' + col for col in features]
)

print("Dataframe filled with KNNImputer: ")
print(mammalKnn_filled.head())

# 3. IterativeImputer - Use 100 iterations
iterative_imputer = IterativeImputer(max_iter=100, random_state=0)
iter_imputed = iterative_imputer.fit_transform(mammalIter)
mammalIter_filled = pd.DataFrame(iter_imputed, columns=features)

print("Dataframe filled with IterativeImputer: ")
print(mammalIter_filled.head())
```
